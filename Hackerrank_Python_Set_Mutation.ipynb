```python
A = int(input())
SET_A = set(map(int, input().split()))
N = int(input())
for i in range(0,N):
    operation = input().split()
    another_set = set(map(int, input().split()))
    if operation[0] == 'update':
       SET_A.update(another_set)
    elif operation[0] == 'intersectioln_update':
       SET_A.intersectioln_update(another_set)
    elif operation[0] == 'symmetric_difference_update':
        SET_A.symmetric_difference_update(another_set)
    elif operation[0] == 'difference_update':
        SET_A.difference_update(another_set)

print(sum(SET_A))
```
